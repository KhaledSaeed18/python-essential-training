## Lists

```python
empty_list = []
```

```python
my_list = [1, 2, 3, 4]
print(my_list)
```

## Sets

```python
my_set = {1, 2, 3, 4}
print(my_set)
```

```python
type(my_set)
```

```python
len(my_set)
```

```python
dup_set = {1, 1, 2, 2}
len(dup_set)
```

```python
print(dup_set)
```

in `Sets` elemets need to be unique, that why we get 2 as length and {1, 2} when print.

```python
{1, 2} == {2, 1} # the order of elemets in sets does not matter
```

```python
{1, 2, 3} == {3, 2, 1, 1, 1} # unique + order does not matter
```

## Dictionaries

```python
my_dictionary = {
    'apple': 'A red fruit', # key:value
    'bear': 'A scary animal'
}
```

```python
my_dictionary['apple']
```

Dictionary `key` must be unique

---

| Data Structure | Symbol / Syntax | Ordered | Mutable | Allows Duplicates | Indexed | Key–Value | Common Use Cases |
|---------------|------------------|---------|---------|-------------------|---------|-----------|------------------|
| **List** | `[ ]` | ✅ Yes | ✅ Yes | ✅ Yes | ✅ Yes | ❌ No | Ordered collection, dynamic data, iteration |
| **Tuple** | `( )` | ✅ Yes | ❌ No | ✅ Yes | ✅ Yes | ❌ No | Fixed data, safety, faster than lists |
| **Set** | `{ }` | ❌ No | ✅ Yes | ❌ No | ❌ No | ❌ No | Uniqueness, membership tests, math ops |
| **Dictionary** | `{ key: value }` | ✅ Yes (Py 3.7+) | ✅ Yes | ❌ Keys only | ❌ No | ✅ Yes | Mappings, configs, fast lookup |
