# Faster Prime Finding Challeneg

Write a function called `allPrimesUpTo` which returns a list of all prime numbers less than the given input

```python
def allPrimesUpTo(num):
    if num <= 2:
        return []

    primesFound = [2]

    for candidate in range(3, num):
        isPrime = True

        for prime in primesFound:
            if prime * prime > candidate:
                break
            if candidate % prime == 0:
                isPrime = False
                break

        if isPrime:
            primesFound.append(candidate)

    return primesFound
```

```python
allPrimesUpTo(100)
```
