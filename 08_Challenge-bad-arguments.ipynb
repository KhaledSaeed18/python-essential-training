# Bad Arguments

Create a custom exception called `handleNonIntArguments`, Fill in the wrapper function in the `handleNonIntArguments`  function to act as an annotation. Check the test code to make sure you understand how this annotation is being used.

```python
class NonIntArgumentsException(Exception):
    pass

def handleNonIntArguments(func):
    def wrapper(*args):
        for item in args:
            if type(item) is not int:
                raise NonIntArgumentsException()
        return func(*args)
    return wrapper
```

```python
@handleNonIntArguments
def sum(a, b, c):
    return a + b + c

try:
    result = sum(1, 2, 'a')
    print('This should not print out')
except NonIntArgumentsException as e:
    print('Hooray!')
```
