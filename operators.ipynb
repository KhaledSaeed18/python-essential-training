# Operators

## Arithmetic Operators

```python
1 + 1 # plus
```

```python
4 * 5 # multiplication
```

```python
5 ** 2 # ^2
```

```python
20 / 5 # division (return float always)
```

```python
20 / 6
```

```python
4 - 2 # minus
```

```python
20 % 6 # modulus (remainder)
```

### Arithmetic Operators with Strings

```python
'string 1 ' + 'string 2'
```

```python
'- string 1 - ' * 4
```

```python
' 1' + 4
```

## Comparison, Logical, and Membership Operators

### Comparison Operators

```python
True == True
```

```python
4 < 5
```

```python
5 <= 5
```

```python
2 > 5
```

### Logical Operators (and, or, not)

```python
True and True
```

```python
True and False
```

```python
True or True
```

```python
False or False
```

```python
not True
```

```python
not False
```

### Membership Operators

```python
1 in [1, 2, 3, 4]
```

```python
6 in [1, 2, 3, 4]
```

```python
10 not in [1, 2, 3, 4]
```

```python
4 not in [1, 2, 3, 4]
```

```python
'cat' in 'my pet cat'
```

```python
'dog' in 'my pet cat'
```
