# Control Flow

## If / Else statements

```python
a = True
if a:
    print('It is true')
    print('Also print this')
print('Always print this')
```

```python
a = False
if a:
    print('It is true')
    print('Also print this')
print('Always print this')
```

```python
a = False
if a:
    print('It is true')
    print('Also print this')
else:
    print('It is false!')
print('Always print this')
```

```python
a = True
b = True
if a:
    print('It is true')
    print('Also print this')
    if b:
        print('Both are true')
else:
    print('It is false!')
print('Always print this')
```

```python
a = True
b = False
c = True
if a:
    print('It is true!')
    print('Also print this')
    if b:
        print('Both are true')
        if c:
            print('All three are true')
else:
    print('It is false!')
print('Always print this')
```

## For loops

```python
a = [1, 2, 3, 4, 5]
for number in a: # number is just a variable name, anything can be used
    print(number)
```

## While loops

```python
a = 0
while a < 5:
    print(a)
    a = a + 1 # increment a to avoid infinite loop
```
