# Functions

```python
print('Hello, world') # Built in function print()
```

```python
def multiplyByThree(val): # define a new function, accept 1 argument
    return 3 * val # the valuer to return when call the function

multiplyByThree(4) # call the function
```

```python
def multiply(val1, val2):
    return val1 * val2
```

```python
multiply(3, 4)
```

```python
a = [1,2,3]

def appendFour(myList): # function without return, function just mutate a list
    myList.append(4)
    
appendFour(a)
print(a)
```
