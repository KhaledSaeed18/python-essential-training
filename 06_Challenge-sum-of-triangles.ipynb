# Sum of triangles

You are given a function triangle that returns the triangular number for a given number num.
A triangular number is the sum of that number and all positive integers less than it.
Fill in the function square that returns the square of a positive integer using only the triangle function.

```python
def triangle(num):
    if num == 1:
        return num
    return num + triangle(num - 1)

def square(num):
    return triangle(num) + triangle(num - 1)
```

```python
square(5)
```

```python
square(6)
```
