# Files

## Reading Files

```python
f = open('10_01-file.txt', 'r')
print(f)
```

```python
f.readline()
```

```python
f.readlines()
```

```python
f = open('10_01-file.txt', 'r')
for line in f.readlines():
    print(line.strip())
```

## Writing Files (Overwrite)

```python
f = open('10_01-output.txt', 'w')
print(f)
```

```python
f.write('Line 1\n')
f.write('Line 2\n')
```

```python
f.close()
```

## Appending Files (Update)

```python
f = open('10_01-output.txt', 'a')
f.write('Line 3\n')
f.write('Line 4\n')
f.close()
```

```python
with open('10_01-output.txt', 'a') as f:
    f.write('some stuff\n')
    f.write('some other stuff\n')
    
print(f)
```
