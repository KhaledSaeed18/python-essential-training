# Lists

## List Slicing

```python
myList = [1,2,3,4,5]
myList[3:] # start at index 3 → go to the end
```

```python
myList[0:6:2] # # start at 0 → stop before 6 → step by 2
```

```python
myList[0:6:3]
```

```python
myList[::2] # start from beginning → go to the end → take every 2nd element
```

```python
myList = list(range(100)) # creates a list from 0 to 99
```

```python
myList[::10] # start from beginning → go to the end → take every 10th element
```

```python
myList[::-10] # start from the end → go backwards → take every 10th element
```

## Modifying Lists

```python
myList = [1,2,3,4]
myList.append(5) # insert at the end
print(myList)
```

```python
myList.insert(3, 'a new value') # inster at index 3, with insert we select index
print(myList)
```

```python
myList.remove('a new value') # remove specific item (not index)
```

```python
print(myList)
```

```python
myList.pop() # remove and return item from the end of the list
```

```python
myList
```

```python
while len(myList):
    print(myList.pop())
```

```python
myList
```

```python
a = [1,2,3,4,5]
b = a
a.append(6)
print(b)
```

```python
a = [1,2,3,4,5]
b = a.copy()
a.append(6)
print(a)
print(b)
```
