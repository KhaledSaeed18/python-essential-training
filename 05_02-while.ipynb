## Break

```python
wait_until = (datetime.now().second + 2) % 60

while True:
    if datetime.now().second == wait_until:
        print(f'We are at {wait_until} seconds!')
        break
```

```python
wait_until = (datetime.now().second + 2) % 60

while True:
    while datetime.now().second == wait_until:
        print(f'We are at {wait_until} seconds!')
        break
```

## Continue

```python
wait_until = (datetime.now().second + 2) % 60

while datetime.now().second != wait_until:
    continue
    print('Still waiting!')
    
    
print(f'We are at {wait_until} seconds!')
```

```python
wait_until = (datetime.now().second + 2) % 60

while True:
    if datetime.now().second < wait_until:
        continue
    break
    
    
print(f'We are at {wait_until} seconds!')
```
