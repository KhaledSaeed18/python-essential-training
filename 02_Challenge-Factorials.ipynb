# Factorial Challenge

**Task:** write a function called factorial that returns the factorial of a number passed in.

```python
def factorial(num):
    if type(num) != int:
        return None
    if num < 0:
        return None
    
    result = 1
    for i in range(1, num + 1):
        result *= i
    return result
```

```python
factorial(5)
```

```python
factorial(0)
```

```python
factorial(-2)
```

```python
factorial(1.2)
```

```python
factorial('spam spam spam spam spam spam')
```
