# Ints and Floats

```python
20 / 4 # division always return float
```

```python
4 + 4.0 # adding to float always return float
```

```python
4 * 4.0
```

```python
4 ** 2.0
```

```python
int(4.5) # bulit in "class", casting the float 4.5 to the integer 4
```

```python
int(8.999999)
```

When casting from float to int python **dont round**

```python
round(8.99999)
```

```python
14/3
```

```python
round(14/3, 2) # 2 is number of decimal places 
```

```python
1.2 - 1.0
```

Computers store numbers in binary (base-2), not decimal (base-10). Some decimal numbers cannot be represented exactly in binary.

```python
round(1.2 - 1.0, 2)
```
