# Classes

```python
class Dog: # Define new class
    def __init__(self, name): # called when crete an instance from this class
        self.name = name # attributes
        self.legs = 4 # attributes
    
    def speak(self): # class method
        print(self.name + ' says: Bark!')
```

```python
my_dog = Dog('Rover') # create new instance from the dog class 
another_dog = Dog('Fluffy')
```

```python
my_dog.speak()
```

```python
another_dog.speak()
```
