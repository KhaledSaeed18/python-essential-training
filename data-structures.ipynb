## Sets

```python
my_set = {1, 2, 3, 4}
print(my_set)
```

```python
type(my_set)
```

```python
len(my_set)
```

```python
dup_set = {1, 1, 2, 2}
len(dup_set)
```

```python
print(dup_set)
```

in `Sets` elemets need to be unique, that why we get 2 as length and {1, 2} when print.

```python
{1, 2} == {2, 1} # the order of elemets in sets does not matter
```

```python
{1, 2, 3} == {3, 2, 1, 1, 1} # unique + order does not matter
```
